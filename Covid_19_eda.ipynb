```python
clean_df = df_country_wise.sort_values(by=["Confirmed"],ascending=False)
clean_df.head(5)
```

```python
clean_df.reset_index(drop = True)
```

```python
px.bar(clean_df[0:10], x="Country/Region", y="Confirmed",title="Top 10 Countries with confirmed cases")
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.barplot(x='Country/Region', y='Deaths', data= clean_df.nlargest(10,'Deaths'))
plt.show()
```

```python
px.pie(clean_df[0:10],values="Confirmed",names="Country/Region")
```

```python
px.bar(clean_df[0:10], x="Country/Region", y="Recovered",title="Top 10 Countries with Recovered cases")
```

```python
px.pie(clean_df[0:10],values= 'Deaths / 100 Cases', names = 'Country/Region',title = 'Average Death Rate')
```

```python
px.bar(data_frame = clean_df[0:10] ,x = 'Country/Region', y = 'Active',title = 'Number of Active Cases ')
```
